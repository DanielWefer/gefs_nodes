```python
import numpy as np
import xarray as xr
import glob
import dask
```

```python
dirs_hgt = glob.glob('/home/scratch/dwefer/GEFSv12/hgt_nc/**/**/*.nc')
dirs_pwat = glob.glob('/home/scratch/dwefer/GEFSv12/pwat_nc/**/**/*.nc')
```

```python
print('len:', len(dirs_hgt), 'same?', len(dirs_hgt) == len(dirs_pwat))
```

```python
datasets = []
for i, dirs in enumerate(dirs_hgt):
    print(f'processing {i + 1}/{len(dirs_hgt)}', end= '\r', flush= True)
    ds = xr.open_dataset(dirs)
    ds.expand_dims(dim= 'time')
    datasets.append(ds)

print('concatenating...')
ds = xr.concat(datasets, dim= 'time')
print('Finished')
```

```python
ds.to_netcdf('/home/scratch/peldridge/hgt_tot.nc')
```

```python
datasets = []
for i, dirs in enumerate(dirs_pwat):
    print(f'processing {i + 1}/{len(dirs_hgt)}', end= '\r', flush= True)
    ds = xr.open_dataset(dirs)
    ds.expand_dims(dim= 'time')
    datasets.append(ds)

print('concatenating...')
ds = xr.concat(datasets, dim= 'time')
print('Finished', flush= True)
```
