```python
import xarray as xr
```

```python
ds = xr.open_dataset("/home/scratch/dwefer/GEFSv12/z500_pwat.nc")
ds
```

```python
gh_mean = ds.gh.mean(dim="time")
gh_std  = ds.gh.std(dim="time")

ds["gh_anomaly"] = (ds.gh - gh_mean) / gh_std
ds["gh_anomaly"].attrs.update({
    "long_name": "Standardized anomaly of geopotential height",
    "description": "z-score relative to mean/std over 20 year dataset",
})


pwat_mean = ds.pwat.mean(dim="time")
pwat_std  = ds.pwat.std(dim="time")

ds["pwat_anomaly"] = (ds.pwat - pwat_mean) / pwat_std
ds["pwat_anomaly"].attrs.update({
    "long_name": "Standardized anomaly of precipitable water",
    "description": "z-score relative to mean/std over 20 year dataset",
})
```

```python
ds.to_netcdf("/home/scratch/dwefer/GEFSv12/z500_pwat_anomalies.nc")
```
